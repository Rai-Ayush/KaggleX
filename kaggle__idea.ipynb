Here we, found that xgboost is giving better results than lgbm but we can use both lgbm and xgb in the form of voting regressor to decide the output this will help us to take insights from both the models and get to equib point...
First we will assign both regressors equal wieghts and then do some experimentation on differnt wieghts ....till now we saw that xgboost is performing better on this dataset so we will try to give it a larger value...🍉🍉

```python
from sklearn.ensemble import VotingRegressor
voting_reg = VotingRegressor(
    estimators=[
        ('lgbm', best_lgbm),
        ('xgb', best_xgb_brand)
    ]
)
voting_reg.fit(X_train, Y_train)
```

```python
output_forset = pd.DataFrame({'Id':dt.id,
                       'price': voting_reg.predict(X_test)})
output_forset.to_csv('submission_vote_lgbm_xgb.csv', index=False)
```

```python
from sklearn.model_selection import cross_val_score

weights_to_test = [[0.1, 0.9], [0.2, 0.8], [0.3, 0.7], [0.4, 0.6], [0.5, 0.5], [0.6, 0.4], [0.7, 0.3], [0.8, 0.2], [0.9, 0.1]]

best_score = float('-inf')
best_weights = None

for weights in weights_to_test:
    voting_reg = VotingRegressor(
        estimators=[
            ('lgbm', best_lgbm),
            ('xgb', best_xgb_brand)
        ],
        weights=weights
    )
    scores = cross_val_score(voting_reg, X_train, Y_train, cv=3, scoring='neg_mean_squared_error')
    mean_score = np.mean(scores)
    if mean_score > best_score:
        best_score = mean_score
        best_weights = weights

voting_reg = VotingRegressor(
    estimators=[
        ('lgbm', best_lgbm),
        ('xgb', best_xgb_brand)
    ],
    weights=best_weights
)
voting_reg.fit(X_train, Y_train)
```

```python
output_forset = pd.DataFrame({'Id':dt.id,
                       'price': voting_reg.predict(X_test)})
output_forset.to_csv('submission_vote_lgbm_xgb_weight.csv', index=False)
```
